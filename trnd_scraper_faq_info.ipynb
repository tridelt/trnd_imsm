```python
# this should work all the way back... 2017 and earlier

df = pd.read_csv(f'./de_projekte.csv')
session = HTMLSession()

csv_infos_line = []
faq_id = 1
for i, row in df.iterrows():
    
    project_base_url = row['href']
#     debugging
#     project_base_url = 'https://www.trnd.com/de/projekte/uncle-bens-mediterran'
    print(f'\n#{i+1}  {project_base_url}')
    homepage = session.get(project_base_url)
    
    nav_bar = homepage.html.find('.project-nav',first=True)
    if nav_bar:
        print([x.text for x in nav_bar.find('a')])
    if nav_bar and any(['Frage' in x.text for x in nav_bar.find('a')]):
        project_faq_link = [x for x in nav_bar.find('a') if 'Frage' in x.text][0].attrs['href']
    elif nav_bar and any(['FAQ' in x.text for x in nav_bar.find('a')]):
        project_faq_link = [x for x in nav_bar.find('a') if 'FAQ' in x.text][0].attrs['href']
    else:
        continue
    
    print(f'#{i+1}  {project_faq_link}')
    faq_page = session.get(project_faq_link)

    faq_elements = faq_page.html.find('.faq-question')
    
    for faq_element in faq_elements:
        q = faq_element.find('h4',first=True)
        a = faq_element.find('p',first=True)
    
        if q and a:
            q = q.text
            a = a.text
    
            csv_infos_line.append({
                'project_id' : row['project_id'],
                'question_id' : faq_id,
                'question' : q,
                'answer' : a
            })
            faq_id+=1
#     break

df = pd.DataFrame(csv_infos_line)
df.to_csv('./de_projekt_faq.csv', index=None)  
```
